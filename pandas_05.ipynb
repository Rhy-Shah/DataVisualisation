```python
df_melt = pd.melt(df,id_vars = ["Date","Drug_Name","Parameter"],
        var_name="Time",
        value_name = "Reading")
df_melt
```

```python
df_pivoted_raw = pd.pivot(df_melt,
         index=["Date","Drug_Name","Time"],
         columns="Parameter",
         values="Reading")
df_pivoted_raw
```

```python
df_pivot.groupby(["Date","Drug_Name"]).groups
```

```python
def calc_mean(x):
    x["temp_mean"] = x["Temperature"].mean()
    x["pressure_mean"] = x["Pressure"].mean()
    return x
```

```python
df_final = df_pivot.groupby(["Date","Drug_Name"]).apply(calc_mean)
df_final
```

```python
df_final["Temperature"].fillna(df_final["temp_mean"],inplace=True)
df_final["Pressure"].fillna(df_final["pressure_mean"],inplace=True)
df_final
```

```python
df_final.drop(columns=["temp_mean","pressure_mean"],inplace=True)
```

```python
df_final
```

```python
temp_bins = [5,20,35,40,60]
temp_labels = ["low","medium","high","very_high"]
```

```python
df_final["temp_cat"] = pd.cut(df_final["Temperature"],
       bins = temp_bins,
       labels = temp_labels)
```
