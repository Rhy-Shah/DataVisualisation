```python
movies=pd.read_csv("movies.csv")
movies.drop(columns="Unnamed: 0",inplace=True)
movies
```

```python
directors = pd.read_csv("directors.csv")
directors.drop(columns="Unnamed: 0",inplace=True)
directors
```

```python
# Find out all the movies created by high budget directors
# High budget director = he has created atleast 1 movies where bugdget was greater than 100 millionn dollars
df.groupby(["director_name"]).filter(lambda x:x["budget"].max()>=100)
```

```python
# Find out all the risky movies
# Risky = Movies with budget more than average revenue of each director
def calculate_risky(x):
    x["risky"] = (x["budget"] - x["revenue"].mean())>=0
    return x
```

```python
df = df.groupby(["director_name"]).apply(calculate_risky)
```
